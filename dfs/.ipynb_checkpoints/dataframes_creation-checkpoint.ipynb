```python
# Imports
import spotipy
import pandas as pd
import json
import pprint
from spotipy.oauth2 import SpotifyClientCredentials
from IPython.display import IFrame
```

```python
# Personal API configuration
import config
```

```python
# Configurate API
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id= config.client_id,
                                                           client_secret= config.client_secret)) 
```

```python
# Useful function
def get_playlist_tracks(username, playlist_id):
    results = sp.user_playlist_tracks(username,playlist_id,market="GB")
    tracks = results['items'] ### In this case, first 100 songs of 288 tracks
    while results['next']: ### While there is a possibility to go further...
        results = sp.next(results) ### ...get the next page...
        tracks.extend(results['items']) ### ...and extend the tracks list with items of the next page
    return tracks
```
