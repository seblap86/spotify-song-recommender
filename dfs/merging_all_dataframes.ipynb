```python
# Importing libraries
import pandas as pd
```

```python
# Importing files

# Decades - 6
decades = [
'1970s.csv',
'1980s.csv',
'1990s.csv',
'2000s.csv',
'2010s.csv',
'2020s.csv'
]

# Genres - Best ofs - 12
main_genres_best_of = [
'best_of-blues_jazz_soul.csv',
'best_of-blues.csv',
'best_of-classic.csv',
'best_of-country.csv',
'best_of_electronic.csv',
'best_of_folk.csv',
'best_of_hiphop.csv',
'best_of_jazz.csv',
'best_of_soul.csv',
'best_of_new_age.csv',
'best_of_reggae.csv',
'best_of_rock.csv'
]

# Other Genres & Subgenres - Best ofs - 28
sub_other_genres_best_of = [
'best_of_afrobeats.csv',
'best_of_britpop.csv',
'best_of_deep_house.csv',
'best_of_dancehall.csv',
'best_of_dreampop.csv',
'best_of_drill.csv',
'best_of_electro_swing.csv',
'best_of_eurodance.csv',
'best_of_grime.csv',
'best_of_house.csv',
'best_of_kpop.csv',
'best_of_latin.csv',
'best_of_melodic_death_metal.csv',
'best_of_metal.csv',
'best_of_neue_deutsche_welle.csv',
'best_of_reggaeton.csv',
'best_of_soundtrack1.csv',
'best_of_soundtrack2.csv',
'best_of_organic_house.csv',
'best_of_pop.csv',
'best_of_punk.csv',
'best_of_rap.csv',
'best_of_rnb.csv',
'best_of_ska.csv',
'best_of_synthwave.csv',
'best_of_trap.csv',
'best_of_trip_hop.csv',
'best_of_world.csv'
]
    
# Some stuff from 2023 - 7
stuff_from_2023 = [
'2023_best_of.csv',
'2023_bravohits.csv',
'2023_deutschrap.csv',
'2023_house_techno.csv',
'2023_the_dome.csv',
'2023_tech_house.csv',
'2023_hiphop_mixtape.csv'
]
 
# Handpicked - 20
handpicked = [
'handpicked_afterlife.csv',
'handpicked_berlin.csv',
'handpicked_be_svendsen.csv',
'handpicked_billie_eilish.csv',
'handpicked_bravo_hits.csv',
'handpicked_bremer_partymix.csv',
'handpicked_faber.csv',
'handpicked_gisbert.csv',
'handpicked_hungry_music.csv',
'handpicked_indie_melody.csv',
'handpicked_karaoke.csv',
'handpicked_manu_chao.csv',
'handpicked_modus_mio.csv',
'handpicked_money_boy.csv',
'handpicked_peter_fox_radio.csv',
'handpicked_peter_licht.csv',
'handpicked_slowdown_tuesday.csv',
'handpicked_sonne_mond_sterne.csv',
'handpicked_the_streets.csv',
'handpicked_worakls.csv'
]
```

```python
decades_dfs = []
for i in decades:
    try:
        decades_dfs.append(pd.read_csv(i))
    except:
        print('No success (decades):', i)
        pass
print('-> decades_dfs:', len(decades_dfs))

main_genres_best_of_dfs = []
for i in main_genres_best_of:
    try:
        main_genres_best_of_dfs.append(pd.read_csv(i))
    except:
        print('No success (main_genres_best_of):', i)
        pass
print('-> main_genres_best_of_dfs:', len(main_genres_best_of_dfs))

sub_other_genres_best_of_dfs = []
for i in sub_other_genres_best_of:
    try:
        sub_other_genres_best_of_dfs.append(pd.read_csv(i))
    except:
        print('No success (sub_other_genres_best_of):', i)
        pass
print('-> sub_other_genres_best_of_dfs:', len(sub_other_genres_best_of_dfs))

stuff_from_2023_dfs = []
for i in stuff_from_2023:
    try:
        stuff_from_2023_dfs.append(pd.read_csv(i))
    except:
        print('No success (stuff_from_2023):', i)
        pass
print('-> stuff_from_2023_dfs:', len(stuff_from_2023_dfs))

handpicked_dfs = []
for i in handpicked:
    try:
        handpicked_dfs.append(pd.read_csv(i))
    except:
        print('No success (handpicked):', i)
        pass
print('-> handpicked_dfs:', len(handpicked_dfs))
```

```python
# All dataframes in one list
all_dfs = []
all_dfs.extend(decades_dfs)
all_dfs.extend(main_genres_best_of_dfs)
all_dfs.extend(sub_other_genres_best_of_dfs)
all_dfs.extend(stuff_from_2023_dfs)
all_dfs.extend(handpicked_dfs)
print('Playlists:', len(all_dfs))

no_songs = 0
for i in all_dfs:
    no_songs += len(i)
print('Songs:', no_songs)
```

```python
# Glueing dataframes together
df = all_dfs[0]
for i in all_dfs[1:(len(all_dfs)+1)]:
    df = pd.concat([df, i])
df.reset_index(drop=True,inplace=True)
df.shape
```

```python
# Checking for and removing duplicates
print('Whole row duplicates:', df.duplicated().sum())
print('Track ID duplicates:', df.duplicated(subset=['id']).sum())

#df.drop_duplicates(inplace=True)
df.drop_duplicates(subset=['id'], inplace=True)
df.reset_index(drop=True, inplace=True)
df.shape
```

```python
# Checking if successful (yes)
print('Whole row duplicates:', df.duplicated().sum())
print('Track ID duplicates:', df.duplicated(subset=['id']).sum())
```
